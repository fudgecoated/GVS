```python
import pandas as pd
```

```python
df = pd.read_csv ('~/Documents/School/GVS/DataCleanUp/DB_NORMALNGVS_JULY242022.csv')
```

```python
pd.set_option('display.max_rows', None)
```

```python
df.columns = ['Time', 'Mode', 'Gain', 'gX', 'gY', 'gZ', 'delete', 'GainGyro', 'Gain Gyro']
```

```python
df.drop(df[df['Mode'] != 'Mode 3 access'].index, inplace = True)
```

```python
df = df.drop(['delete', 'Gain Gyro'],axis = 1)
#df = df.drop([0])
```

```python
#df = df.drop([], axis=0, inplace=True)
```

```python
df['gX'] = df["gX"].str.replace("gX:", "")
df['gY'] = df["gY"].str.replace("gY:", "")
df['gZ'] = df["gZ"].str.replace("gZ:", "")
df['GainGyro'] = df['GainGyro'].str.replace("Gain", "")
df['GainGyro'] = df['GainGyro'].str.replace("*", "")
df['GainGyro'] = df['GainGyro'].str.replace("gyroY:", "")
```

```python
df['Mode'] = df["Mode"].str.replace("Mode ", "")
df['Mode'] = df["Mode"].str.replace(" access", "")
```

```python
#df.drop(df[df['gYZ'] == 'NaN'].index, inplace = True)
```

```python
df[['Date','Time']] = df.Time.str.split(" ",expand=True,)
```

```python
df = df[["Date", "Time", "Gain", "gX", "gY" , "gZ" , "GainGyro"]]
```
